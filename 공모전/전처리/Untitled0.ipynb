```python
#속도데이터를 모두 가져와서 합치고 권역구분이 광진구이며 차선수가 2 이상인 데이터
#만 남겨서 결측치 제거함함
speed1=pd.read_excel(r'2022년 3월 서울시 차량통행속도.xlsx')
speed2=pd.read_excel(r'2022년 4월 서울시 차량통행속도.xlsx')
speed3=pd.read_excel(r'2022년 5월 서울시 차량통행속도.xlsx')
speed4=pd.read_excel(r'2022년 6월 서울시 차량통행속도.xlsx')
speed5=pd.read_excel(r'2022년 7월 서울시 차량통행속도.xlsx')
speed6=pd.read_excel(r'2022년 8월 서울시 차량통행속도.xlsx')
speed8=pd.read_excel(r'2022년 9월 서울시 차량통행속도.xlsx')
speed7=pd.read_excel(r'2022년 10월 서울시 차량통행속도.xlsx')
speed9=pd.read_excel(r'2022년 11월 서울시 차량통행속도.xlsx')
speed10=pd.read_excel(r'2022년 12월 서울시 차량통행속도.xlsx')
speed11=pd.read_excel(r'2023년 1월 서울시 차량통행속도.xlsx')
speed12=pd.read_excel(r'2023년 2월 서울시 차량통행속도.xlsx')


speed = pd.concat([speed1, speed2, speed3, speed4, speed5, speed6, speed7,speed8, speed9])

speed_data = pd.concat([speed, speed10, speed11, speed12])

speed_data = speed_data.dropna()

speed_data = speed_data[speed_data['권역구분']=='광진구']

speed_data = speed_data[speed_data['차선수']>=2]

print(speed_data)
```
