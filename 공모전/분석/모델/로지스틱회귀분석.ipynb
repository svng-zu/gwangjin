```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/공모전/데이터/학습용/데이터변환2.csv")
print(data.shape)
print(data.head())
```

```python
x = data[['시간','교통량','요일', '방향', '거리', '차선수', '기능유형구분', '합류도로수', '제한속도', '주말 유무']]
y = data[['속도']]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_features = scaler.fit_transform(x_train)
test_features = scaler.transform(x_test)
```

```python
test_features.astype('float')
```

```python
y_train.astype('int')
y_test.astype('int')
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(train_features, y_train.astype('int'))
```

```python
print(model.score(train_features, y_train.astype('int')))
```

```python
print(model.score(test_features, y_test.astype('int')))
```

```python
print(model.coef_)
```
